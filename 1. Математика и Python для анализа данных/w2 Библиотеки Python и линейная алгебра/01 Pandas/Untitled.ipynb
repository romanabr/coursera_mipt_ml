## Интерполяция функции экспоненты

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(0, 20, 2)
y = np.exp(-x/3)
z = x ** 2 / 100

print('x = ', x)
print('y = ', y)
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

```python
import scipy
scipy.__version__
```

```python
f = interpolate.interp1d(x, y)
print('f= ', f)
x_new = np.arange(0, 8, 0.1)
y_new = f(x_new)
plt.plot(x, y, 'o')
plt.plot(x_new, y_new, '-')
```
