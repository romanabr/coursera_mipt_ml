Рассмотрим сложную математическую функцию на отрезке [1, 15]:

$ f(x) = sin(\frac{x}{5})*e^{\frac{-x}{2}} $


```python
import numpy as np
import math
from matplotlib import pylab as plt
%matplotlib inline

def orig_function(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
x = np.arange(1, 15.1, 0.1)
y = orig_function(x)

print(len(x), x[0], x[len(x)-1])
print(len(x), y[0], y[len(x)-1])
plt.plot(x, y)
plt.show()
```

### Многочлен первой степени
Сформируйте систему линейных уравнений (то есть задайте матрицу коэффициентов A и свободный вектор b) для многочлена первой степени, который должен совпадать с функцией f в точках 1 и 15. Решите данную систему с помощью функции scipy.linalg.solve. Нарисуйте функцию f и полученный многочлен. Хорошо ли он приближает исходную функцию?
Система уравнений первой степени:

$ f(x) = w_0 + w_1*x_1 $

$ 3.25221687 = w_0 + w_1 $

$ 0.63522142 = w_0 + w_1 * 15$

$$ \begin{vmatrix} 1 & 1 \\ 1 & 15 \end{vmatrix} * \begin{vmatrix} x_1 \\ x_2 \end{vmatrix} = \begin{vmatrix} 3.25221687 \\ 0.63522142 \end{vmatrix}$$

```python
matrix1 = np.array([[1,1],[1,15]])
result1 = np.array([3.25221687, 0.63522142])
x1 = np.linalg.solve(matrix1, result1)
print("x1: ",x1)

def func1(x):
    return 3.43914512 - 0.18692825*x

x = np.arange(1, 15.1, 0.1)
plt.plot(x, orig_function(x))
plt.plot(x, func1(x))
plt.show()
```

### Многочлен второй степени
Повторите те же шаги для многочлена второй степени, который совпадает с функцией f в точках 1, 8 и 15. Улучшилось ли качество аппроксимации?

$ f(x) = w_0 + w_1*x + w_2*x^2$


```python
print(orig_function(np.array([1, 8, 15])))
```

$ 3.25221687 = w_0 + w_1 + w_2$

$ 2.31617016 = w_0 + 8 w_1 + 64w_2$

$ 0.63522142 = w_0 + 15w_1  + 225w_2$

$$ \begin{vmatrix} 1 & 1 & 1 \\ 1 & 8& 64 \\ 1 & 15& 225\end{vmatrix} * \begin{vmatrix} x_1 \\ x_2 \\ x_3 \end{vmatrix} = \begin{vmatrix} 3.25221687 \\2.31617016 \\ 0.63522142 \end{vmatrix}$$

```python
matrix2 = np.array([[1,1,1],[1,8, 64], [1,15,225]])
result2 = np.array([3.25221687, 2.31617016, 0.63522142])
x2 = np.linalg.solve(matrix2, result2)
print("x2: ",x2)

def func2(x):
    return 3.3251295 - 0.06531159*x - 0.00760104 * x**2

x = np.arange(1, 15.1, 0.1)
plt.plot(x, orig_function(x))
plt.plot(x, func1(x))
plt.plot(x, func2(x))
plt.show()





```

### Многочлен третей степени
Повторите те же шаги для многочлена третьей степени, который совпадает с функцией f в точках 1, 4, 10 и 15. Хорошо ли он аппроксимирует функцию? Коэффициенты данного многочлена (четыре числа в следующем порядке: w_0, w_1, w_2, w_3) являются ответом на задачу. Округлять коэффициенты не обязательно, но при желании можете произвести округление до второго знака (т.е. до числа вида 0.42)

$ f(x) = w_0 + w_1*x + w_2*x^2 + w_3*x^3$

```python
print(orig_function(np.array([1, 4, 10, 15])))
```

$ 3.25221687 = w_0 + w_1 + w_2 + w_3$

$ 1.74684595 = w_0 + 4 w_1 + 16w_2+ 64w_3$

$ 2.50541641 = w_0 + 10w_1  + 100w_2+ 1000w_3$

$ 0.63522142 = w_0 + 15w_1  + 225w_2+ 3375w_2$

$$ \begin{vmatrix} 1 & 1 & 1& 1 \\ 1 & 4& 16 & 64 \\ 1 & 10& 100 & 1000\\ 1 & 15& 225 & 3375\end{vmatrix} * \begin{vmatrix} x_1 \\ x_2 \\ x_3 \\x_4 \end{vmatrix} = \begin{vmatrix} 3.25221687 \\1.74684595 \\ 2.50541641 \\ 0.63522142 \end{vmatrix}$$


```python
matrix3 = np.array([[1,1,1,1],[1,4,16, 64],[1,10,100,1000], [1,15,225, 3375]])
result3 = np.array([3.25221687, 1.74684595, 2.50541641, 0.63522142])
x3 = np.linalg.solve(matrix3, result3)
print("x3: ",x3)

def func3(x):
    return 4.36264154 -1.29552587*x + 0.19333685 * x**2 -0.00823565*x**3
# 4.36264154 -1.29552587  0.19333685 -0.00823565
x = np.arange(1, 15.1, 0.1)
plt.plot(x, orig_function(x))
plt.plot(x, func1(x))
plt.plot(x, func2(x))
plt.plot(x, func3(x))
plt.show()
```
